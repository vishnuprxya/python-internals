```python
# ==========================================
# ðŸš— Car Price Analyzer â€” Complete Visualization Project
# ==========================================

# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# STEP 1: Create Random Car Dataset
# -------------------------------
np.random.seed(42)  # for reproducibility

brands = ['Maruti', 'Hyundai', 'Honda', 'Toyota', 'Ford', 'Tata', 'Mahindra', 'Kia', 'Renault', 'Skoda']
fuel_types = ['Petrol', 'Diesel', 'CNG']
transmissions = ['Manual', 'Automatic']

data = {
    'Brand': np.random.choice(brands, 50),
    'Year': np.random.randint(2010, 2023, 50),
    'Fuel_Type': np.random.choice(fuel_types, 50),
    'Transmission': np.random.choice(transmissions, 50),
    'Kms_Driven': np.random.randint(10000, 150000, 50),
    'Selling_Price': np.random.uniform(2.0, 15.0, 50).round(2)
}

df = pd.DataFrame(data)

print("ðŸ”¹ Sample Car Data (first 10 rows):\n")
print(df.head(10))

# Save dataset (optional)
df.to_csv("car_data.csv", index=False)
print("\nâœ… Dataset saved as 'car_data.csv'")

# -------------------------------
# STEP 2: Summary Statistics
# -------------------------------
print("\nðŸ“Š Summary Statistics:\n")
print(df.describe())

# -------------------------------
# STEP 3: Data Visualizations
# -------------------------------
sns.set(style="whitegrid", palette="muted")

# 1ï¸âƒ£ Distribution of Selling Price
plt.figure(figsize=(7,4))
sns.histplot(df['Selling_Price'], kde=True, color='royalblue', bins=10)
plt.title('Distribution of Car Selling Prices')
plt.xlabel('Selling Price (in Lakhs)')
plt.ylabel('Frequency')
plt.show()

# 2ï¸âƒ£ Average Price by Brand
plt.figure(figsize=(10,5))
sns.barplot(x='Brand', y='Selling_Price', data=df, estimator='mean', ci=None, palette='viridis')
plt.title('Average Car Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Selling Price (Lakhs)')
plt.xticks(rotation=45)
plt.show()

# 3ï¸âƒ£ Car Price vs Year (with Regression Line)
plt.figure(figsize=(7,4))
sns.regplot(x='Year', y='Selling_Price', data=df, scatter_kws={'s':60}, line_kws={'color':'red'})
plt.title('Car Price vs Year (Regression Trend)')
plt.xlabel('Manufacturing Year')
plt.ylabel('Selling Price (Lakhs)')
plt.show()

# 4ï¸âƒ£ Fuel Type Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Fuel_Type', data=df, palette='Set2')
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()

# 5ï¸âƒ£ Transmission Type Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Transmission', data=df, palette='Set3')
plt.title('Manual vs Automatic Cars')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.show()

# 6ï¸âƒ£ Correlation Heatmap (Numerical Features)
plt.figure(figsize=(6,4))
corr = df[['Year', 'Kms_Driven', 'Selling_Price']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

# 7ï¸âƒ£ Pairplot to Explore Feature Relationships
sns.pairplot(df, vars=['Year', 'Kms_Driven', 'Selling_Price'], hue='Fuel_Type', palette='husl', height=2.5)
plt.suptitle('Pairplot: Relationship Between Year, Kms, and Price', y=1.02)
plt.show()

print("\nâœ… All Visualizations Generated Successfully!")
```
